```python
# import bibliotek
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation, NMF
from nltk.corpus import PlaintextCorpusReader
from nltk.tokenize import word_tokenize
import json
import string
import morfeusz2
import itertools as it
import operator as op
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# utworzenie korpusu dokumentów
corpus_dir = "./Literatura - streszczenia - oryginal"
corpus = PlaintextCorpusReader(corpus_dir, ".*\.txt")
files =corpus.fileids()
print(*files, sep="\n")
```

```python
# wczytanie dokuemntów z korpusu do słownika (dictionary)
documents = {}
for file in files:
    documents[file.split(".txt")[0]] = corpus.raw(file)
print(json.dumps(documents, indent=4, ensure_ascii=False))
```

```python
# wczytanie stopisty z pliku
with open("./stopwords_pl.txt", "r", encoding="UTF-8") as stoplist_file:
    stoplist = stoplist_file.read().splitlines()
print(*stoplist, sep="\n")
```

```python
# definicja funkcji do lematyzacji autorstwa Prof. UEK Pawła Wołoszyna
def lemmatize(text):
    morf = morfeusz2.Morfeusz()
    segments = it.groupby(morf.analyse(text),op.itemgetter(0,1))
    def disambiguate(group):
        pairs = ((len(descr),lemma)
        for _, _, (_, lemma, descr, _, _,) in group)
        perpl, lemma = min(pairs)
        return lemma.split(":")[0]
    lemmas = (disambiguate(group) for key, group in segments)
    return " ".join(filter(str.isalpha, lemmas))
```

```python
# wstępne przygotowanie dokumentów
for key in documents:
    documents[key] = documents[key].lower()
    documents[key] = "".join([char for char in documents[key] if char not in string.digits])
    documents[key] = "".join([char for char in documents[key] if char not in string.punctuation])
    documents[key] = lemmatize(documents[key])
    documents[key] = " ".join([word for word in word_tokenize(documents[key]) if word not in stoplist])
print(json.dumps(documents, indent=4, ensure_ascii=False))
```

```python
# wczytanie przetworzonych dokumentów se słownika do ramki danych (DataFrame)
docs = pd.DataFrame.from_dict(documents, orient="index")
docs.columns = ['content']
docs
```

```python
# utworzenie macierzy częstości ważonej tf
count_vectorizer = CountVectorizer()
counts_tf = count_vectorizer.fit_transform(docs['content'])
counts_tf
```

```python
# utworzenie macierzy częstości ważonej tfidf
tfidf_vectorizer = TfidfVectorizer()
counts_tfidf = tfidf_vectorizer.fit_transform(docs['content'])
counts_tfidf
```

```python
# konwersja macierzy rzadkiej do macierzy klasycznej
counts_tf_m = counts_tf.toarray()
counts_tfidf_m = counts_tfidf.toarray()
counts_tfidf_m
```

```python
# parametry modelu topic modellong
topics_count = 4
words = count_vectorizer.get_feature_names_out()
display_words_count = 20
subplots = [1, 4]
```

```python
# źródło https://scikit-learn.org/stable/auto_examples/applications/plot_topics_extraction_with_nmf_lda.html
# definicja funkcji do rysowania wykresów tematów
def plot_top_words(model, feature_names, n_top_words, title):
    fig, axes = plt.subplots(*subplots, figsize=(30, 15), sharex=True)
    axes = axes.flatten()
    for topic_idx, topic in enumerate(model.components_):
        top_features_ind = topic.argsort()[:-n_top_words - 1:-1]
        top_features = [feature_names[i] for i in top_features_ind]
        weights = topic[top_features_ind]

        ax = axes[topic_idx]
        ax.barh(top_features, weights, height=0.7)
        ax.set_title(f'Temat {topic_idx +1}',
                     fontdict={'fontsize': 30})
        ax.invert_yaxis()
        ax.tick_params(axis='both', which='major', labelsize=20)
        for i in 'top right left'.split():
            ax.spines[i].set_visible(False)
        fig.suptitle(title, fontsize=40)

    plt.subplots_adjust(top=0.90, bottom=0.05, wspace=0.90, hspace=0.3)
    plt.show()
```

```python
# zdefiniowanie modelu LDA (Latent Dirichlet Allocation)
lda = LatentDirichletAllocation(
    n_components=topics_count,
    max_iter=5, 
    learning_method='online',
    learning_offset=50.,
    random_state=0
)
```

```python
# uczenie modelu LDA
lda.fit(counts_tf)
```

```python
# rysowanie wykresów ważności słów w tematach
plot_top_words(lda, words, display_words_count, "Tematy w modelu LDA")
```

```python
# zdefiniowanie modelu NMF (Non-negative Matrix Factorization) - Frobenius norm
nmf_f = NMF(
    n_components=topics_count, 
    random_state=1, 
    alpha=.1,
    l1_ratio=.5
) 
```

```python
# uczenie modelu NMF - Frobenius norm
nmf_f.fit(counts_tfidf)
```

```python
# rysowanie wykresów ważności słów w tematach
plot_top_words(nmf_f, words, display_words_count, "Tematy w modelu NMF (Morma macierzowa)")
```

```python
# zdefiniowanie modelu NMF (Non-negative Matrix Factorization) - generalized Kullback-Leibler divergence
nmf_kl = NMF(
    n_components=topics_count, 
    random_state=1, 
    beta_loss='kullback-leibler',
    solver="mu",
    max_iter=1000,
    alpha=.1,
    l1_ratio=.5
) 
```

```python
# uczenie modelu NMF - Frobenius norm
nmf_kl.fit(counts_tfidf)
```

```python
# rysowanie wykresów ważności słów w tematach
plot_top_words(nmf_kl, words, display_words_count, "Tematy w modelu NMF (Uogulniona dywergencja Kullback'a-Leibler'a)")
```
